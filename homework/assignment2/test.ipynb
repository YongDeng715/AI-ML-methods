```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import numpy as np

X_train_path = 'Iris/train/x.txt'
y_train_path = 'Iris/train/y.txt'
X_test_path = 'Iris/test/x.txt'
y_test_path = 'Iris/test/y.txt'

def make_dataset(X_path, y_path):
    x_ori = np.loadtxt(X_path)
    x_ori_std = np.std(x_ori, axis=0)
    x_ori_mean = np.mean(x_ori, axis=0)
    x_tmp = (x_ori - x_ori_mean) / x_ori_std

    X_aug = np.concatenate([np.ones((x_tmp.shape[0], 1)), x_tmp], axis=1)
    y = np.loadtxt(y_path)
    y = y.reshape(len(y), 1)

    return X_aug, y

X_train, y_train = make_dataset(X_train_path, y_train_path)
X_test, y_test = make_dataset(X_test_path, y_test_path)
```

```python
reg = LogisticRegression(multi_class='multinomial')
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
print("Logistic Regression model accuracy(in %):",\
      metrics.accuracy_score(y_test, y_pred)*100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
np.random.seed(0)
num_samples = 200
num_classes = 3
X = 2 * np.random.rand(num_samples, 2) - 1
y = np.random.randint(num_classes, size=num_samples)

# One-hot 
def one_hot_encode(y, num_classes):
    one_hot = np.zeros((len(y), num_classes))
    one_hot[np.arange(len(y)), y] = 1
    return one_hot

y_one_hot = one_hot_encode(y, num_classes)

#  softmax 
def softmax(logits):
    exp_logits = np.exp(logits - np.max(logits, axis=1, keepdims=True))
    return exp_logits / np.sum(exp_logits, axis=1, keepdims=True)

# 
def cross_entropy_loss(y_true, y_pred):
    return -np.sum(y_true * np.log(y_pred + 1e-10)) / len(y_true)

# 
num_features = X.shape[1]
learning_rate = 0.01
num_epochs = 1000

W = np.random.randn(num_features, num_classes)
b = np.zeros((1, num_classes))

# 
losses = []
for epoch in range(num_epochs):
    logits = np.dot(X, W) + b
    y_pred = softmax(logits)
    loss = cross_entropy_loss(y_one_hot, y_pred)
    losses.append(loss)

    # 
    grad_logits = (y_pred - y_one_hot) / len(y_one_hot)
    grad_W = np.dot(X.T, grad_logits)
    grad_b = np.sum(grad_logits, axis=0, keepdims=True)
    W -= learning_rate * grad_W
    b -= learning_rate * grad_b


    #  100 
    if epoch % 100 == 0:
        plt.figure(figsize=(8, 6))    
        plt.title(f'Epoch {epoch}, Loss: {loss:.4f}')
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
        xx, yy = np.meshgrid(np.linspace(-1.5, 1.5, 100), np.linspace(-1.5, 1.5, 100))
        grid = np.c_[xx.ravel(), yy.ravel()]
        probs = softmax(np.dot(grid, W) + b)
        preds = np.argmax(probs, axis=1)
        plt.contourf(xx, yy, preds.reshape(xx.shape), cmap=plt.cm.Spectral, alpha=0.8)
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.colorbar()
        plt.show()

# 
plt.figure(figsize=(8, 6))
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```
